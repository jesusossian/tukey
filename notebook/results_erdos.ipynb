```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="erdos"
method_="lp"

form_="form0"

file=f"/home/jossian/repository/tukey/result/{method_}/{method_}_tukey_{inst_}_{form_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum()
    },index=["lp_karate_form0"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_erdos_form0 = table

#print(lp_erdos_form0[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
inst_="erdos"
method_="lp"

form_="form1"

file=f"/home/jossian/repository/tukey/result/{method_}/{method_}_tukey_{inst_}_{form_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum()
    },index=["lp_karate_form0"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_erdos_form1 = table

#print(lp_erdos_form1[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
inst_="erdos"
method_="lp"

form_="indForm1"

file=f"/home/jossian/repository/tukey/result/{method_}/{method_}_tukey_{inst_}_{form_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum()
    },index=["lp_karate_form0"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_erdos_cutInd = table

#print(lp_erdos_cutInd[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
inst_="erdos"
method_="lp"

form_="rltForm1"

file=f"/home/jossian/repository/tukey/result/{method_}/{method_}_tukey_{inst_}_{form_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum()
    },index=["rlt_karate"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_erdos_rlt = table

#print(lp_erdos_rlp[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
result = pd.DataFrame()
result = pd.merge(
    lp_erdos_form0, lp_erdos_form1,
    on=['vertice'], 
    suffixes=("", "_f1"),
    )

result = pd.merge(
    result, lp_erdos_cutInd,
    on=['vertice'], 
    suffixes=("", "_ind"),
    )

result = pd.merge(
    result, lp_erdos_rlt,
    on=['vertice'], 
    suffixes=("", "_rlt"),
    )

result

#print(result[['vertice','bb','bb_f1','bb_ind','bb_rlt']].to_latex(index=False,float_format="%.1f"))

with open('../result/table.tex', 'w') as tf:
     tf.write(
          result[[
               'vertice',
               'bb','bb_f1','bb_ind','bb_rlt'
               ]].to_latex(index=False,float_format="%.1f")
               )
```
