```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="random_internet_100"
method_="lp"

form_="form0"

file=f"/home/jossian/repository/tukey/result/{method_}/{method_}_tukey_{inst_}_{form_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum()
    },index=["lp_karate_form0"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_internet_form0 = table

print(lp_internet_form0[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```
