```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="karate"
method_="lp"
ver_="fmax"
```

```python
form_="fmaxc0"

file=f"../result/{inst_}/{method_}/{ver_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum(),
    'tk':table['bb'].max()
    },index=[f"{inst_}"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_karate_c0 = table
#lp_karate_c0

#print(lp_karate_c0[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```
