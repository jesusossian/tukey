```python
G = nx.Graph()


with open("../instances/Cuneiform/Cuneiform_A.txt", "r") as filestream:
        for line in filestream:
            currentline = line.split(",")
            G.add_edge(int(currentline[0]),int(currentline[1]))

M = G.number_of_edges()
M = G.number_of_nodes()

G.edges
```
