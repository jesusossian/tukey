```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="internet_graph"
method_="lp"
dim_=200
id_=1
```

```python
form_="fminc1"
ver_="fmin"

file=f"../result/{inst_}/{method_}/{ver_}/{method_}_{form_}_{inst_}_{dim_}_{id_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}_{form_}_{dim_}_{id_}",
    'bb':table["bb"].sum(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum(),
    'tk':table['bb'].max()
    },index=[f"{form_}_{inst_}_{dim_}_{id_}"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_internet_fminc1 = table
lp_internet_fminc1

#print(lp_internet_fminc1[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```
