```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="karate"
method_="exact"

file=f"/home/jossian/repository/tukey/result/{inst_}_{method_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bb':table["bb"].max(),
    'bs':table['bs'].max(),
    'gap':table['gap'].mean(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].max(),
    'opt':table['opt'].sum()
    },index=["karate"]
)

table = pd.concat([table, resume], ignore_index=True)
exact = table

print(exact[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
inst_="karate"
method_="rf"

file=f"/home/jossian/repository/tukey/result/{inst_}_{method_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bs','time']

resume = pd.DataFrame({
    'vertice':f"{inst_}",
    'bs':table['bs'].max(),
    'time':table['time'].sum(),
    },index=["rf"]
)

table = pd.concat([table, resume], ignore_index=True)
rf = table

print(rf[['vertice','bs','time']].to_latex(index=False,float_format="%.2f"))
```

```python
result = pd.DataFrame()
result = pd.merge(
    exact, rf, 
    on=['vertice'], 
    suffixes=("_exact", "_rf"),
    )

result
```
