```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="karate"
method_="lp"
```

```python
form_="fmin"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum(),
    'tk':table['bb'].max()
    },index=[f"{inst_}"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_karate_fmin = table
lp_karate_fmin

#print(lp_karate_fmin[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
form_="minmiset"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum(),
    'tk':table['bb'].max()
    },index=[f"{inst_}"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_karate_minmiset = table
lp_karate_minmiset

#print(lp_karate_minmiset[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
form_="maxmiset"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','time','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'time':table['time'].sum(),
    'opt':table['opt'].sum(),
    'tk':table['bb'].max()
    },index=[f"{inst_}"]
)

table = pd.concat([table, resume], ignore_index=True)
lp_karate_maxmiset = table
lp_karate_maxmiset

#print(lp_karate_maxmiset[['vertice','bb','time','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
result = pd.DataFrame()

result = pd.merge(
    lp_karate_fmin, lp_karate_fmax,
    on=['vertice'], 
    suffixes=("", "_max"),
    )

result = pd.merge(
    result, lp_karate_minmiset,
    on=['vertice'], 
    suffixes=("", "_minmiset"),
    )

result = pd.merge(
    result, lp_karate_maxmiset,
    on=['vertice'], 
    suffixes=("", "_maxmiset"),
    )

#result
result[['vertice','bb','tk','bb_max','tk_max','bb_minmiset','tk_minmiset','bb_maxmiset','tk_maxmiset']]
```
