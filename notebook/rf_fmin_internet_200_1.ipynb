```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="internet_graph"
method_="rf"
dim_=200
id_=1
ver_="fmin"
```

```python
form_="fminc1"

file=f"../result/{inst_}/zeta/{method_}_{form_}_{inst_}_{dim_}_{id_}_3_2.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bs','time']

resume = pd.DataFrame({
    'vertice':f"{method_}_{form_}_{dim_}_{id_}",
    'bs':table["bs"].sum(),
    'time':table['time'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}_{inst_}_{dim_}_{id_}"]
)

table = pd.concat([table, resume], ignore_index=True)
rf_internet_c1_3_2 = table
rf_internet_c1_3_2

#print(mip_internet_c0[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```
