```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="karate"
method_="mip"
```

```python
form_="fmin"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'bs':table["bs"].sum(),
    'gap':table['gap'].sum(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].sum(),
    'opt':table['opt'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}"]
)

table = pd.concat([table, resume], ignore_index=True)
mip_karate_fmin = table
mip_karate_fmin

#print(mip_karate_fmin[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
form_="fmax"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'bs':table["bs"].sum(),
    'gap':table['gap'].sum(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].sum(),
    'opt':table['opt'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}"]
)

table = pd.concat([table, resume], ignore_index=True)
mip_karate_fmax = table
mip_karate_fmax

#print(mip_karate_fmax[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
form_="maxmiset"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'bs':table["bs"].sum(),
    'gap':table['gap'].sum(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].sum(),
    'opt':table['opt'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}"]
)

table = pd.concat([table, resume], ignore_index=True)
mip_karate_maxmiset = table
mip_karate_maxmiset

#print(mip_internet_maxmiset_200_1[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
form_="minmiset"

file=f"../result/{inst_}/{method_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'bs':table["bs"].sum(),
    'gap':table['gap'].sum(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].sum(),
    'opt':table['opt'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}"]
)

table = pd.concat([table, resume], ignore_index=True)
mip_karate_minmiset = table
mip_karate_minmiset

#print(mip_karate_minmiset[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```

```python
result = pd.DataFrame()

result = pd.merge(
    mip_karate_fmin, mip_karate_fmax,
    on=['vertice'], 
    suffixes=("", "_fmax"),
    )

result = pd.merge(
    result, mip_karate_minmiset,
    on=['vertice'], 
    suffixes=("", "_minmiset"),
    )

result = pd.merge(
    result, mip_karate_maxmiset,
    on=['vertice'], 
    suffixes=("", "_maxmiset"),
    )

result[[
    'vertice',
    'bs','gap','time','opt','tk','tmax',
    'bs_fmax','gap_fmax','time_fmax','opt_fmax','tk_fmax','tmax_fmax',
    'bs_minmiset','gap_minmiset','time_minmiset','opt_minmiset','tk_minmiset','tmax_minmiset',
    'bs_maxmiset','gap_maxmiset','time_maxmiset','opt_maxmiset','tk_maxmiset','tmax_maxmiset'
    ]]
```
