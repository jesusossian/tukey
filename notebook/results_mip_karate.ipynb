```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inst_="karate"
method_="mip"
ver_="fmin"
```

```python
form_="fminc1"

file=f"../result/{inst_}/{method_}/{ver_}/{method_}_{form_}_{inst_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['vertice','bb','bs','gap','time','nodes','opt']

resume = pd.DataFrame({
    'vertice':f"{method_}",
    'bb':table["bb"].sum(),
    'bs':table["bs"].sum(),
    'gap':table['gap'].sum(),
    'time':table['time'].sum(),
    'nodes':table['nodes'].sum(),
    'opt':table['opt'].sum(),
    "tk":table["bs"].max(),
    "tmax":table["time"].max()
    },index=[f"{form_}"]
)

table = pd.concat([table, resume], ignore_index=True)
mip_karate_fminc1 = table
mip_karate_fminc1

#print(mip_karate_fminc1[['vertice','bb','bs','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```
